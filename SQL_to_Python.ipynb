Notebook Objectives
* Converting data to pandas dataframes
* Pickling dataframes for quick access

```python
from sqlalchemy import create_engine
import pandas as pd
import datetime
```

```python
cnx = create_engine('postgresql://ubuntu@3.23.86.219:5432/flights')

# weather data for ATL airport
weather = pd.read_sql_query(
    '''
    SELECT * FROM airport_weather aw WHERE airportcode = 'KATL'
    ''', cnx
)

weather.info()
```

```python
month_mask = weather['StartTime(UTC)'].map(lambda x: x.month) == 1
year_mask = weather['StartTime(UTC)'].map(lambda x: x.year) == 2018

weather_jan18 = weather[month_mask & year_mask]
```

```python
weather_jan18.to_pickle('/Users/wasilaq/Metis/Proj3/pickled_files/weather_jan18')
```

```python
# flight data for ATL airport
flights = pd.read_sql_query(
    '''
    SELECT * FROM flights_jan18 fj WHERE origin = 'ATL'
    ''', cnx
)

flights.head()
```
